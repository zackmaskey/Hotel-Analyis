```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Read file 
file_path=r"C:\Users\Samrachhika\Downloads\Hotel Magh 2079.xlsx"
df = pd.read_excel(file_path, sheet_name='Magh',header=2)#Sheet name is magh and column header starts from 3rd row
```

```python
#Check for null values and replace them with 0

print(df.columns.tolist())#PRINTS COLUMNS
#select columns that might contain null values 
columns_to_replace = ['QTY', 'UOM','CREDIT','CASH','REMARKS']
#replace null values with 0
df[columns_to_replace] = df[columns_to_replace].fillna(0)




```

```python
df.head(50)
```

```python
#Income rows filter
income_df = df[df['INCOME/EXPENSE'] == 'Income']

# Group by DATE and calculate totals for CASH and CREDIT
daily_sales = income_df.groupby('DATE')[['CASH', 'CREDIT']].sum()

print(daily_sales)

# Plot the data
plt.figure(figsize=(14, 4))  
plt.plot(daily_sales.index, daily_sales['CASH'], label='Cash Sales', marker='o', color='blue')
plt.plot(daily_sales.index, daily_sales['CREDIT'], label='Credit Sales', marker='s', color='green')

# Add labels, title, and legend
plt.xlabel('Date')
plt.ylabel('Sales Total')
plt.title('Daily Sales: Cash vs Credit')
plt.legend()

plt.grid(True)


plt.show()
```

```python
# Group by DATE and INCOME/EXPENSE type to calculate total CASH and CREDIT
daily_sales_expenses = df.groupby(['DATE', 'INCOME/EXPENSE'])[['CASH', 'CREDIT']].sum()

print(daily_sales_expenses)

# Data for stacked bar chart
fig, ax = plt.subplots(figsize=(10,4))
for label, group in daily_sales_expenses.groupby('INCOME/EXPENSE'):
    ax.bar(
        group.index.get_level_values(0),  # Dates
        group['CASH'] + group['CREDIT'],  # Sum of Cash and Credit
        label=label
    )

# Add labels, title, and legend
plt.xlabel('Date')
plt.ylabel('Amount')
plt.title('Stacked Sales and Expenses')
plt.legend()
plt.show()

```

```python
#Total sales, total cash sales and total credit sales
total_cash_sales=income_df['CASH'].sum()
total_credit_sales=income_df['CREDIT'].sum()
total_sales= total_cash_sales+total_credit_sales
print(f"Total Cash sales: {total_cash_sales}")
print (f"Total Credit Sales: {total_credit_sales}")
print(f"Total Sales: {total_sales}")

```

```python
#Income generating department
department_sales=income_df.groupby('DEPARTMENT')[['CASH', 'CREDIT']].sum().sort_values(by=['CASH', 'CREDIT'], ascending=False)
# new column 'Total' which is the sum of 'CASH' and 'CREDIT'
department_sales['Total'] = department_sales['CASH'] + department_sales['CREDIT']

print(department_sales)

#Plotting Pie chart
labels = department_sales.index  # Department names
sizes = department_sales['Total']  # Total income for each department
colors = plt.cm.tab20.colors[:len(department_sales)]  # Use a colormap for distinct colors

# Plot the pie chart
plt.figure(figsize=(6,5))
plt.pie(
    sizes,
    labels=labels,
    autopct='%1.1f%%',  # Display percentage
    startangle=90,      # Rotate for better alignment
    colors=colors,
    wedgeprops={'edgecolor': 'black'}  # Add border for clarity
)

# Add title
plt.title('Income Distribution by Department')

# Display the chart
plt.tight_layout()
plt.show()
```

```python
department_expenses = expense_df.groupby('DEPARTMENT')[['CASH', 'CREDIT']].sum()

# Add a 'Total' column for the sum of CASH and CREDIT
department_expenses['Total'] = department_expenses['CASH'] + department_expenses['CREDIT']

print(department_expenses)

# plotting pie chart
labels = department_expenses.index  # Department names
sizes = department_expenses['Total']  # Total expenses for each department
colors = plt.cm.tab20.colors[:len(department_expenses)]  # Generate distinct colors

# Plot the pie chart
plt.figure(figsize=(6,4))
plt.pie(
    sizes,
    labels=labels,
    autopct='%1.1f%%',  # Display percentage
    startangle=60,   
    colors=colors,
    wedgeprops={'edgecolor': 'black'}  # Add borders for clarity
)
plt.title('Expense Distribution by Department')

# Display the chart
plt.tight_layout()
plt.show()
```

```python
# Filter for Expense rows and Administration department
admin_expenses = df[(df['INCOME/EXPENSE'] == 'Expense') & (df['DEPARTMENT'] == 'Adminstration')]

# Group by TYPE and calculate total expenses (CASH + CREDIT)
admin_expenses_breakdown = admin_expenses.groupby('TYPE')[['CASH', 'CREDIT']].sum()

# Add a 'Total' column for the sum of CASH and CREDIT
admin_expenses_breakdown['Total'] = admin_expenses_breakdown['CASH'] + admin_expenses_breakdown['CREDIT']

# Sort by 'Total' in descending order for better clarity
admin_expenses_breakdown = admin_expenses_breakdown.sort_values(by='Total', ascending=False)

# Display the breakdown
print(admin_expenses_breakdown)

# Plot a pie chart 
plt.figure(figsize=(8,8))
admin_expenses_breakdown['Total'].plot(kind='pie', labels=admin_expenses_breakdown.index, autopct='%1.1f%%', startangle=45, colors=colors,wedgeprops={'edgecolor': 'black'})

# Add title
plt.title('Proportion of Total Expenses by Type in Administration Department')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Filter for Expense rows and Administration department
kitchen_expenses = df[(df['INCOME/EXPENSE'] == 'Expense') & (df['DEPARTMENT'] == 'Kitchen')]

# Group by TYPE and calculate total expenses (CASH + CREDIT)
kitchen_expenses_breakdown = kitchen_expenses.groupby('TYPE')[['CASH', 'CREDIT']].sum()

# a 'Total' column for the sum of CASH and CREDIT
kitchen_expenses_breakdown['Total'] =kitchen_expenses_breakdown['CASH'] + kitchen_expenses_breakdown['CREDIT']


print(kitchen_expenses_breakdown)

# Plot the pie chart
plt.figure(figsize=(8, 8))
kitchen_expenses_breakdown['Total'].plot(kind='pie', autopct='%1.1f%%', startangle=70, colors=colors,wedgeprops={'edgecolor': 'black'})

# Title and other plot formatting
plt.title('Kitchen Expenses Breakdown')
plt.ylabel('')  # Hide y-axis label for clarity


plt.show()
```

```python
#Kitchen Expenses VS Restaurant Income
# Filter for Restaurant Income
restaurant_income = df[(df['INCOME/EXPENSE'] == 'Income') & (df['DEPARTMENT'] == 'Resturant')]

# Group by DATE and sum the CASH and CREDIT for Kitchen Expenses
kitchen_expenses_daily = kitchen_expenses.groupby('DATE')[['CASH', 'CREDIT']].sum()

# Group by DATE and sum the CASH and CREDIT for Restaurant Income
restaurant_income_daily = restaurant_income.groupby('DATE')[['CASH', 'CREDIT']].sum()

# Add a 'Total' column for each (CASH + CREDIT) in both dataframes
kitchen_expenses_daily['Total'] = kitchen_expenses_daily['CASH'] + kitchen_expenses_daily['CREDIT']
restaurant_income_daily['Total'] = restaurant_income_daily['CASH'] + restaurant_income_daily['CREDIT']

# Merge the data on 'DATE' to compare Kitchen Expense vs Restaurant Income on the same day
comparison_df = pd.merge(kitchen_expenses_daily[['Total']], restaurant_income_daily[['Total']], 
                         left_index=True, right_index=True, suffixes=('_Kitchen', '_Restaurant'))

print(comparison_df)

#plot bar graph

comparison_df.plot(kind='bar', figsize=(12, 6))

# Add labels and title
plt.xlabel('Date')
plt.ylabel('Amount')
plt.title('Kitchen Expenses vs Restaurant Income')

# Show the plot
plt.tight_layout()
plt.show()
```
